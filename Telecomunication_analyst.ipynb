# Data Preprocessing

### **IMPORT**

```python
import pandas as pd
import numpy as np
```

### **Read Dataset**

```python
telco = pd.read_excel('https://raw.githubusercontent.com/rezanagita/telecommunication-analytics/main/dataset%20telco.xlsx')
telco.head()
```

```python
telco.info()
```

```python
telco.describe()
```

### **Rename column**

```python
# Rename column Area = City, Cluster = Province
telco.rename(columns={'Area': 'City', 'Cluster': 'Province'}, inplace=True)
telco.head()
```

```python
telco.rename(columns={'Tenure_Month': 'Tenure_Days'}, inplace=True)
```

### **Ubah value type data: Activation_Date**

```python
# Mengubah Tipe data ke datetime
telco['Activation_Date'] = pd.to_datetime(telco['Activation_Date'], format='mixed', dayfirst=True)
telco.info()
```

### **Merapikan value activation date**
ganti 2026 menjadi 2025

```python
latest_date= telco['Activation_Date'].max()
latest_date
```

```python
# mengganti tahun 2026 menjadi 2025
# Karena 'Activation_Date' adalah tipe datetime, kita gunakan
telco['Activation_Date'] = telco['Activation_Date'].apply(lambda x: x.replace(year=2025) if x.year == 2026 else x)
telco.head()
```

```python
latest_date= telco['Activation_Date'].max()
latest_date
```

### **Merapikan nama City dan Province**

```python
# Melihat value unik di city
telco['City'].unique()
```

```python
# Merapikan value City
city_mapping = {
    'Jogja': 'Yogyakarta',
    'SBY': 'Surabaya',
    'Bandng': 'Bandung',
    'JAKARTA': 'Jakarta',
    'Jakrta': 'Jakarta',
    'JKT': 'Jakarta'
}
telco['City'] = telco['City'].replace(city_mapping)
telco.head()
```

```python
# Melihat value unik di Province
telco['Province'].unique()
```

```python
telco['Province'] = telco['City'].map(
    {
        'Yogyakarta': 'DIY',
        'Surabaya': 'Jawa Timur',
        'Bandung': 'Jawa Barat',
        'Jakarta': 'DKI'
    }
)
telco.head()
```

### **Merapikan value Channel**

```python
# Melihat value unik di Channel
telco['Channel'].unique()
```

```python
# ubah dealer ke Dealer
telco['Channel'] = telco['Channel'].replace('dealer', 'Dealer')
```

```python
# Melihat value unik di Channel
telco['Channel'].unique()
```

### **Handling tenure_month**

```python
# menghilangkan tanda minus
telco['Tenure_Days'] = telco['Tenure_Days'].abs()
telco.head()
```

### **Handle Missing Value**

```python
telco.isnull().sum()
```

```python
telco['Discount_Pct'].isna().sum()
```

```python
telco['Discount_Pct'].isna().mean() * 100
```

```python
# fill discount_pict
telco['Discount_Pct'].fillna(0, inplace=True)
```

```python
telco.isnull().sum()
```

```python
# handle missing Subscriber_ID fill dengan unknown
telco['Subscriber_ID'].fillna('Unknown', inplace=True)
```

### **Feature Engineering**

```python
# menambahkan kolom profit
telco['Profit'] = telco['Recharge_Amount'] - telco['Network_Cost']
telco.head()
```

```python
# menambahkan Kategori NPS_Score
# kategori < 51 buruk
# kategori >= 51 - 85 bagus
# kategori >= 85 bagus sekali

telco['NPS_Category'] = np.where(telco['NPS_Score'] < 51, 'Buruk',
np.where((telco['NPS_Score'] >= 51) & (telco['NPS_Score'] <= 85), 'Bagus', 'Bagus Sekali'))
telco.head()
```

```python
# ARPU_Gap --> selisih pendapatan aktual (Recharge_Amount) vs target (Target_ARPU)
telco['ARPU_Gap'] = telco['Recharge_Amount'] - telco['Target_ARPU']

# kategorikan cat_target_ARPU jika positif = mencapai (Y) dan negatif = tidak mencapai (N)
def cat_target_ARPU (x) :
  if x <= 0 :
    return 'N'
  else :
    return 'Y'
telco['cat_target_ARPU'] = telco['ARPU_Gap'].apply(cat_target_ARPU)
telco.head()
```

```python
# Menampilkan baris di mana cat_target_ARPU bernilai 'Y'
telco[telco['cat_target_ARPU'] == 'Y'].head()
```
